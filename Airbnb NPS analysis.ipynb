# Detractors & Promoters

- We need to calculate NPS(Net Promoter Score)
- **NPS = % Promoters - % Detractors**
- To calculate Promoters -> count of surveys having Score>=9.
- To calculate Detractors -> count of surveys having Score<=6.



```python
# Let's get the count of promoters and detractors.

# Promoter count
promoters=data[data>=9].shape[0]

# Detractor count
detractors=data[data<=6].shape[0]
```

```python
promoters
```

```python
detractors
```

```python
# Calculating % Promoters
per_promot=promoters/total*100

per_promot
```

```python
# Calculating % Detractors
per_det=detractors/total*100

per_det
```

## Calculating NPS

```python
nps=np.round(per_promot-per_det)
nps
```
